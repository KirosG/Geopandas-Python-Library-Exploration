# Animal Strikes Data Cleaning

While working on a map for understanding animal behavior I was exploring some data from VCGI, specifically the animal strikes dataset from 2006. My first attempt at mapping it made me realized that the data needed some cleanup for my needs. Specifically I wanted to understand what kind of animals were being hit, and manually going through fields in QGIS did not make sens. In the following sections I will import, find and replace values, verify assumptions and then export the data into a new data file.

```python
animalstrikes = gpd.read_file('VT_VehicleAnimal_Collisions__2006.shp')
```

```python
summarypre = animalstrikes[['MSRI_CODE','DATE_']].groupby(['MSRI_CODE'], as_index=False).count()
summarypre.columns = ['Type','Count']
summarypre['Perc'] = summarypre['Count'] / summarypre['Count'].sum()
summarypre
```

A quick way to understand what needs to be corrected, it is obvious that Other contains 13% and needs to be address. Below I also expore Otter because later on I realized a lot were being reasigned by my process from otter to other types. It looks like some of these may have meant to type in "Other" rather than "Otter" and we will try to rectify that.  The other surpise is that moose account for 52% of strikes.  Perhaps this is an inherient bias as smaller animals are likely not reported.

```python
otter = animalstrikes[(animalstrikes['MSRI_CODE'] == 'Otter')]
otter = otter[['OBJECTID', 'MSRI_CODE', 'MSRI_DESCR','COMMENT']]
otter.head(35)
```

## Finding and Replacing Values

The next sequence is going through and replacing values either literally or looking for the word contained in the comments column. This is when I discovered that Otter was having some specific issues

```python
animalstrikes['MSRI_CODE'] = animalstrikes['MSRI_CODE'].str.replace("Deer RK AB","Deer")
animalstrikes['MSRI_CODE'] = animalstrikes['MSRI_CODE'].str.replace("Deer RK AD","Deer")
animalstrikes['MSRI_CODE'] = animalstrikes['MSRI_CODE'].str.replace("Deer RK UN","Deer")
animalstrikes['MSRI_CODE'] = animalstrikes['MSRI_CODE'].str.replace("Deer RK FD","Deer")
animalstrikes['MSRI_CODE'] = animalstrikes['MSRI_CODE'].str.replace("Deer RK FB","Deer")
```

```python
pd.unique(animalstrikes['MSRI_CODE'])
```

```python
animalstrikes['COMMENT'] = animalstrikes['COMMENT'].str.lower()
```

```python
other = animalstrikes[(animalstrikes['MSRI_CODE'] == 'Other')]
other = other[['OBJECTID', 'MSRI_CODE', 'MSRI_DESCR','COMMENT']]
print(other.shape)
other.head(10)
```

```python
animalstrikes.loc[animalstrikes["COMMENT"] == "coyote", "MSRI_CODE"] = 'Coyote'
animalstrikes.loc[animalstrikes["COMMENT"] == "woodchuck", "MSRI_CODE"] = 'Woodchuck'
animalstrikes.loc[animalstrikes["COMMENT"] == "muskrat", "MSRI_CODE"] = 'Muskrat'
animalstrikes.loc[animalstrikes["COMMENT"] == "skunk", "MSRI_CODE"] = 'Skunk'
animalstrikes.loc[animalstrikes["COMMENT"] == "raccoon", "MSRI_CODE"] = 'Raccoon'
```

```python
pd.unique(animalstrikes['MSRI_CODE'])
```

```python

animalstrikes.loc[(animalstrikes['COMMENT'].str.contains("coyote")) & (animalstrikes['COMMENT'].notnull()), "MSRI_CODE"] = 'Coyote'
animalstrikes.loc[(animalstrikes['COMMENT'].str.contains("rabbit")) & (animalstrikes['COMMENT'].notnull()), "MSRI_CODE"] = 'Rabbit'
animalstrikes.loc[(animalstrikes['COMMENT'].str.contains("raccoon")) & (animalstrikes['COMMENT'].notnull()), "MSRI_CODE"] = 'Raccoon'
animalstrikes.loc[(animalstrikes['COMMENT'].str.contains("coon")) & (animalstrikes['COMMENT'].notnull()), "MSRI_CODE"] = 'Raccoon'
animalstrikes.loc[(animalstrikes['COMMENT'].str.contains("bear")) & (animalstrikes['COMMENT'].notnull()), "MSRI_CODE"] = 'Bear'
animalstrikes.loc[(animalstrikes['COMMENT'].str.contains("deer")) & (animalstrikes['COMMENT'].notnull()), "MSRI_CODE"] = 'Deer'
animalstrikes.loc[(animalstrikes['COMMENT'].str.contains("bobcat")) & (animalstrikes['COMMENT'].notnull()), "MSRI_CODE"] = 'Bobcat'
animalstrikes.loc[(animalstrikes['COMMENT'].str.contains("dog")) & (animalstrikes['COMMENT'].notnull()), "MSRI_CODE"] = 'Dog'
animalstrikes.loc[(animalstrikes['COMMENT'].str.contains("beaver")) & (animalstrikes['COMMENT'].notnull()), "MSRI_CODE"] = 'Beaver'
animalstrikes.loc[(animalstrikes['COMMENT'].str.contains("fox")) & (animalstrikes['COMMENT'].notnull()), "MSRI_CODE"] = 'Fox'
animalstrikes.loc[(animalstrikes['COMMENT'].str.contains("skunk")) & (animalstrikes['COMMENT'].notnull()), "MSRI_CODE"] = 'Skunk'
```

```python
other = animalstrikes[(animalstrikes['MSRI_CODE'] == 'Other')]
other = other[['OBJECTID', 'MSRI_CODE', 'MSRI_DESCR','COMMENT']]
print(other.shape)
other.head(45)
```

```python
animalstrikes = animalstrikes[(animalstrikes['DATE_'].notnull())]
animalstrikes['DATE'] = pd.to_datetime(animalstrikes['DATE_'])
#animalstrikes['timeonly'],animalstrikes['dateonly']= animalstrikes['DATE'].apply(lambda x:x.time()), animalstrikes['DATE'].apply(lambda x:x.date())
animalstrikes.head(50)
```

```python
animalstrikes.info()
```

```python
animalstrikes['MONTH'] = animalstrikes['DATE'].dt.month
animalstrikes['DAY'] = animalstrikes['DATE'].dt.day
animalstrikes['YEAR'] = animalstrikes['DATE'].dt.year

animalstrikes.head()
```

```python
data = animalstrikes[(animalstrikes['MSRI_CODE'] == 'Moose')]
sns.boxplot(x='YEAR', y='MONTH', data=data)
plt.title('Distribution of Animal Strikes for Moose')
```

```python
data = animalstrikes[(animalstrikes['MSRI_CODE'] == 'Deer')]
sns.boxplot(x='YEAR', y='MONTH', data=data)
plt.title('Distribution of Animal Strikes for Deer')
```

```python
data = animalstrikes[(animalstrikes['MSRI_CODE'] == 'Bear')]
sns.boxplot(x='YEAR', y='MONTH', data=data)
plt.title('Distribution of Animal Strikes for Bear')
```

```python
data = animalstrikes[(animalstrikes['MSRI_CODE'] == 'Bear') | (animalstrikes['MSRI_CODE'] == 'Moose') | (animalstrikes['MSRI_CODE'] == 'Deer')]
data = data[(data['YEAR'] > 2000)]
sns.violinplot(x='YEAR', y='MONTH', hue='MSRI_CODE', data=data)
plt.title('Distribution of Animal Strikes for 2001-2006')
```

```python
data = animalstrikes[(animalstrikes['YEAR'] == 2005)]
sns.boxplot(x='YEAR', y='MONTH', hue='MSRI_CODE', data=data)
plt.title('Distribution of Animal Strikes for 2004')
plt.ylim(1,12)
```

```python
corr = animalstrikes
MSRI = pd.get_dummies(animalstrikes['MSRI_CODE'] , prefix = 'MSRI')
corr = pd.concat([corr, MSRI], axis=1)

corr = corr.corr()
sns.heatmap(corr)
plt.show()
```

```python
# Shapefiles do not know what to do with a datetime format so it needs to be converted back to a string for exporting
animalstrikes['DATE'] = str(animalstrikes['DATE'])
# Verify that is worked
animalstrikes.info()
```

## Creating New Shapefile

After doing all that work I want to export the updated data as a new shapefile. It is a simple command and then the file is ready to be consumed. In this case I then exported it back into QGIS for visualization.

```python
# Determine the output path for the Shapefile
out = r"animalstrikes2006.shp"

# Write the data into that Shapefile
animalstrikes.to_file(out)
```
